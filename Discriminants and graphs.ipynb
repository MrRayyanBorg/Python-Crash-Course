# Discriminants 

### We know the $x$-intercepts of a quadratic graph $$y=ax^{2}+bx+c$$ can be found using the quadratic formula: $$x=\frac{-b\pm\sqrt{b^{2}-4ac}}{2a}$$

### The **discriminant** of a quadratic equation is:  $$\vartriangle=b^{2}-4ac$$

# Possible conclusions from Geogebra activity:

### Task 1

#### Observation 1: The graph moves position. When it does not touch or cross the $x$-axis the discriminant is negative.

#### Observation 2: When the discriminant is negative, the graph has no roots on $y=0$. When the discriminant is $0$ the graph only has $1$ root on $y=0$.


### Task 2

#### Fix $b$ and $c$. Suppose we say $b=2$ and $c=7$. Then $$\vartriangle=4-28a$$

#### If the discriminant is negative then we have $$4-28a<0$$<p> $$\implies a>\frac{4}{28}=\frac{1}{7}$$

#### This means for $a>\frac{1}{7}$ we have no roots!

#### For $a=\frac{1}{7}$ we have a single root. 

#### For $a=0$, our equation is actually linear. For all other values we have two distinct roots!

#### Extra observation that could be useful later: The graph changes orientation and width as you vary $a$.

#### We find that we have a root on $c=\frac{9}{8}$. For $c>\frac{9}{8}$  our quadratic has no roots.

#### Extra observation that could be useful later: The graph moves up and down along the $y$-axis as you change $c$.

### Task 4

#### Fix $a$ and $c$. For example suppose $a=2$ and $c=5$. Our discriminant is then<p> $$\vartriangle=b^{2}-40$$

#### Can we find the values of $b$ for which this is true?

#### An interesting thing to do is work out the discriminant for a few values of $b$ and plot them on a set of axes.

```python
import toyplot
import numpy
import math

# define the discriminant as a function that is dependent on b
def f(b):
    return b**2-40
    f(b)
```

```python
# Make a list of b-values
bs = [-20,-17.5,-15,-12.5,-10,-7.5,-2*math.sqrt(10),-5,-2.5,0,2.5,5,2*math.sqrt(10),7.5,10,12.5,15,17.5,20]
# For each b-value, D=f(b)
Ds = [f(b) for b in bs]
Ds
```

```python
canvas, axes, marks = toyplot.scatterplot(bs,Ds,size=10,width=400)
axes.x.spine.position = 0
axes.y.spine.position = 0
```

#### We find that the determinant is negative when $b$ is in between the two roots of $D=b^{2}-40=0$. What are these two roots? 
#### $$b=\pm2\sqrt{10}$$
#### This means that $y$ does *not* have roots when $b$ is in the range $$-2\sqrt{10}<b<2\sqrt{10}$$
